# Machine Learning Project: Pima Indians

In this project I am going to investigate independent variables in the Pima Indian Dataset (blood pressure, skin thickness, etc) and evaluate their effectiveness in predicting whether a member of the dataset has diabetes. 

### Loading The Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

data = pd.read_csv(url, names=names)
data.head()
```

```python
# describe basic aspects of the data
description = data.describe() 
description
```

### Visualising Raw Data

```python
from pandas.plotting import scatter_matrix

# visualise data in histograms and scatter plots relative to each other.
scatter_matrix(data, figsize=(7,7))
plt.show()
```

### Pre-Processing The Data

```python
from sklearn.preprocessing import StandardScaler

# get data values
array = data.values

# split into X and Y values
X = array[:, :-1]
Y = array[:, -1]

# standardise the data
scaler = StandardScaler().fit(X)
rescaledX = scaler.fit_transform(X)

# summarise the new data
np.set_printoptions(precision=4)

print(rescaledX[0:5,:])
```

### Evaluation of Resampling Methods

Comparing the accuracy of a model trained with basic Train/Test sets against a model trained and tested through k-fold cross validation. 

The k-fold cross validation proved more accurate for this dataset. 

```python
from sklearn.model_selection import KFold, train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression

kfold = KFold(n_splits=10, random_state=7)
lm = LogisticRegression(solver='liblinear')

# evaluate cross validation score
results_acc = cross_val_score(lm, X, Y, cv=kfold)
print(("Cross Validation Accuracy: %.3f%% (%.3f%%)") % (results_acc.mean()*100.0, results_acc.std()*100.0))

# evaluate basic train/test split score
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
model = lm.fit(X_train, y_train)
print("Train/Test Split Accuracy:", "%.3f%%" % (model.score(X_test, y_test)*100))
```

### Using Negative Log Loss for Algorithm Evaluation

```python
# evaluate negative log loss
scoring = 'neg_log_loss'
results_nll = cross_val_score(lm, X, Y, cv=kfold, scoring=scoring)
print(("Negative Log Loss: %.3f (%.3f)") % (results_nll.mean(), results_nll.std()))
```

### Visualising Confusion Matrix and Classification Report

```python
from sklearn.metrics import classification_report, confusion_matrix

# generate predicted classes from test set
y_pred = model.predict(X_test)

# compare predicted classes to actual classes and generate accuracy report
report = classification_report(y_test, y_pred, target_names=['Non-Diabetic', 'Diabetic'])
print(report)
```

The model generates a very respectable F1 score of 0.78. F1 score is important in particular to this dataset, as we want to minimise the number of False Positives and False Negatives (i.e. misdiagnosing patients).
